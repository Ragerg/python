```python
import matplotlib.pyplot as plt
```

```python
plt.title('plot test')
plt.plot([10, 20, 30 ,40])
```

```python
plt.plot([10, 20, 30, 40],[12, 45, 23, 87]) # plt.plot([x축 데이터],[y축 데이터])
```

```python
plt.plot([10, 20, 30 ,40], label='asc')
plt.plot([40, 30, 20, 10], label='desc')
plt.legend()
```

```python
plt.plot([10, 20, 30 ,40], label='asc')
plt.plot([40, 30, 20, 10], label='desc')
plt.legend(loc=6)
```

```python
plt.plot([10, 20, 30 ,40], label='asc', color='red', linestyle='dashed') # linestyle = ls
plt.plot([40, 30, 20, 10], 'blue', label='desc', ls = 'dotted') # 속성값을 쓰는게 뒤로
plt.legend(loc=9)
```

```python
plt.plot([10, 20, 30 ,40], label='asc', color='r', linestyle='--')
plt.plot([40, 30, 20, 10], 'g', label='desc', ls = ':') # solid -' / dashed'--' / dotted':' / dash-dot'-.'
plt.legend(loc=7)
```

```python
plt.plot([10, 20, 30, 40], 'r.')
plt.plot([40, 30, 20, 10], 'b^', ls='-.')
```
