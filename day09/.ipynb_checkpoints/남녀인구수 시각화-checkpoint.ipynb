```python
import csv
import matplotlib.pyplot as plt
```

```python
# 방법1
with open('data/202308_gender.csv','r') as file:
    data = csv.reader(file)
    header = next(data) # 제목없애고
    town = '성복'
    print(header[3], header[3+100], header[106], header[106+100])
    for row in data: # 데이터 한 줄 씩
        if town in row[0]:
            male = list()
            female = list()
            for i in range(0, 101):
                male.append(int(row[i+3].replace(',','')))
                female.append(int(row[i+106].replace(',','')))
    print(male)
    print(female)
```

```python
# 남성인구수를 음수로 변환
for i in range(len(male)):
    male[i] = -male[i]
print(male)
```

```python
plt.barh(range(101), male)
plt.barh(range(101), female)
```

```python
# 방법 2
with open('data/202308_gender.csv','r') as file:
    data = csv.reader(file)
    next(data)
    town = '성복'
    for row in data:
        if town in row[0]:
            male = [int(i.replace(',', '')) for i in row[3:104]]
            female = [int(i.replace(',', '')) for i in row[106:]]
    print(male)
    male = [-i for i in male]
    print(male)
```

```python
# 시각화
plt.rc('font', family='gulim')
plt.title(f'{town}동 연령별 남/여 그래프')
plt.barh(range(101), male, label='남')
plt.barh(range(101), female, label='여')
plt.legend()
```

```python
# 남녀비율 원그래프 시각화
plt.pie([-sum((male)), sum(female)], labels=['남자', '여자'], startangle=90, autopct='%.1f%%')
plt.legend()
```

```python
male = [-i for i in male] if male[0]<0 else male
totalCnt = [male[i]+female[i] for i in range(len(male))]
plt.scatter(male, female, s =totalCnt, c = range(len(male)), alpha=0.5)
plt.colorbar()
plt.xlabel('남자인구수')
plt.ylabel('여자인구수')
```
