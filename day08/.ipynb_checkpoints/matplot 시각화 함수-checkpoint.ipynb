```python
import matplotlib.pyplot as plt
# 막대그래프(세로)
xlist = [1, 10, 20, 30, 40]
ylist = [5, 20, 14, 50 ,36]
plt.bar(xlist, ylist)
```

```python
# 막대그래프(가로)
plt.barh(xlist, ylist)
```

```python
# 히스토그램(빈도수)
data = [10, 1, 20, 20, 40, 1, 10, 30, 40, 30, 30, 1, 1, 10]
plt.hist(data, bins = 40)
```

```python
# 주사위를 10번 던져 1 ~ 6의 숫자의 빈도수 시각화
import random
numbers = []
for i in range(1000000):
    numbers.append(random.randint(1,6))
    
plt.hist(numbers, bins = 6)
```

```python
numbers = []
for i in range(11):
    numbers.append(random.randint(1, 10000))
print(sorted(numbers))

plt.boxplot(numbers)
```

```python
plt.pie([10, 30, 20], labels=['test1', 'test2', 'test3'], autopct='%.1f%%')
plt.legend()
```
