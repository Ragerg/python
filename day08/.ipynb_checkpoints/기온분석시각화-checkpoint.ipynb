```python
# 가장 더운 날 구하기
with open('data/seoul-2022.csv', 'r') as file:
    data = csv.reader(file)
    highTemplist = list()
    next(data) # 첫 번째 행은 패스
    for row in data:
        # print(raw[-1])
        if len(row) > 1 and row[-1] != '':
            highTemplist.append(float(row[-1]))
    print(highTemplist)        
    print('최고온도 : ', max(highTemplist)) 
    print(len(highTemplist))    
```

```python
import matplotlib.pyplot as plt

with open('data/seoul-2022.csv', 'r') as file:
    data = csv.reader(file)
    highTemplist = list()
    next(data)
    for row in data:
        # print(raw[-1])
        if len(row) > 1 and row[-1] != '':
            highTemplist.append(float(row[-1]))
            
    plt.figure(figsize = (10, 3))        
    plt.plot(highTemplist, 'r')
    plt.show()
```

```python
# 매년 8월의 최고온도를 시각화
with open('data/seoul-1970and2023.csv', 'r') as file:
    data = csv.reader(file)
    next(data)
    highTemplist = []
    for row in data:
        if len(row) >1 and row[-1] != '':
            date = row[0].split('-')
            month = date[1]
            if month == '08':
                # print(row)
                highTemplist.append(float(row[-1]))
                
    plt.figure(figsize=(10, 3))
    plt.plot(highTemplist, 'r')
```

```python
# 매년 내 생일의 최고, 최저온도를 시각화
with open('data/seoul-1970and2023.csv', 'r') as file:
    highTemp = []
    lowTemp = []
    years = []
    
    data = csv.reader(file)
    next(data)
    for row in data:
        if len(row) >1 and row[-1] != '':
            ymd = row[0].split('-')
            if ymd[1] == '03' and ymd[2] == '25':
                highTemp.append(float(row[-1]))
                lowTemp.append(float(row[-2]))
                years.append(ymd[0].strip())
                
    plt.figure(figsize=(15, 3))
    plt.xticks(rotation=90)
    plt.plot(years, highTemplist, 'r')
    plt.plot(years, lowTemplist, 'b')
```
