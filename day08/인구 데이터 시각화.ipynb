```python
# 경기도 성남시 수정구 (4113100000) 연령별 인구수 막대그래프로 시각화
with open('data/ageTotal_202308.csv', 'r') as file:
    data = csv.reader(file)
    ageList = []
    
    for row in data:
        if row[0]  == '경기도 성남시 수정구 (4113100000)':
#             print(row)
            for age in  row[3:]:
                ageList.append(int(age.replace(',', '')))
            
    plt.bar(range(101), ageList)
    plt.plot(ageList)
```
